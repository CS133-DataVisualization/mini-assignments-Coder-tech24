#### Question 3: ####
How many times does each country appeared in this dataset? Show each country and the corresponding count (show counts in ascending order)

```python
import pandas as pd
csvurl = "https://gist.githubusercontent.com/clairehq/79acab35be50eaf1c383948ed3fd1129/raw/407a02139ae1e134992b90b4b2b8c329b3d73a6a/winemag-data-130k-v2.csv"
wine = pd.read_csv(csvurl)

# Drop the first column (index)
wine.drop(wine.columns[0], axis="columns", inplace=True)

# Count how many times each country appears and sort in ascending order
country_counts = wine['country'].value_counts().sort_values()

# Print the results
print(country_counts) # in a dataframe
```

#### Question 4: ####
Create a variable `adjusted_price` containing the adjusted price which is the price subtracted by the average price. *This is called **"centering" transformation** - a method commonly used in the preprocessing step before applying various machine learning algorithms.*

```python
# average price
average_price = wine['price'].mean()

# find the adjusted price
wine['adjusted_price'] = wine['price'] - average_price

# Print the first few rows.
print(wine[['price', 'adjusted_price']].head())
```

#### Question 5: ####
What is the title of the wine that has the highest points-to-price ratio in the dataset?

```python
# points to price ratio
wine['points_to_price_ratio'] = wine['points'] / wine['price']

# Find the wine name with the highest ratio
highest_ratio_wine = wine.loc[wine['points_to_price_ratio'].idxmax(), 'title']

print(highest_ratio_wine)
```

#### Question 6: ####
Create a series `flavor_counts` that contains two values: the number of wines that has the word "tart" in the `description` column and the number of wines that has the word "berries" in the `description` column. The index of the Series should be "Tart" and "Berries" for the corresponding values.

```python
# Count the number of wines with "tart" in the description
tart_count = wine['description'].str.contains('tart', case=False, na=False).sum()

# Count the number of wines with "berries" in the description
berries_count = wine['description'].str.contains('berries', case=False, na=False).sum()

# Create the flavor_counts Series
flavor_counts = pd.Series([tart_count, berries_count], index=['Tart', 'Berries'])

# Display the Series
print(flavor_counts)
```

#### Question 7: ####
Let's convert the points into simple star ratings. A score of 90 or higher counts as 3 stars, a score of at least 80 but less than 90 is 2 stars. Any other score is 1 star.

Also, any wines from France should automatically get 3 stars, regardless of points.

Add this new column `star_ratings` to the dataframe with the number of stars for each wine in the dataset.

```python
import numpy as np

# Define a function to determine star ratings
def get_star_rating(row):
    if row['country'] == 'France':
        return 3
    elif row['points'] >= 90:
        return 3
    elif 80 <= row['points'] < 90:
        return 2
    else:
        return 1

# Apply the function to create the star_ratings column
wine['star_ratings'] = wine.apply(get_star_rating, axis=1)

# Display the first few rows to verify
print(wine[['points', 'country', 'star_ratings']].head())
```

#### Question 9: ####
What combination of countries and varieties are most common? Create a Series whose index is a MultiIndexof {country, variety} pairs. For example, a pinot noir produced in the US should map to {"US", "Pinot Noir"}. Sort the values in the Series in descending order based on wine count.

```python
# Group by country and variety, then count the occurrences
country_variety_counts = wine.groupby(['country', 'variety']).size()

# Convert the counts to a Series with a MultiIndex
country_variety_series = country_variety_counts.sort_values(ascending=False)

# Display the Series
print(country_variety_series)
```

#### Question 10 #####
Create a Series whose index is reviewers and whose values is the average score given out by that reviewer. Hint: you will need the `taster_name` and `points` columns.

```python
# Group by taster_name and calculate the average score
average_scores = wine.groupby('taster_name')['points'].mean()

# Convert the result to a Series
average_scores_series = average_scores.sort_values(ascending=False)

# Display the Series
print(average_scores_series)
```
